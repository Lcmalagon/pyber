```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
import pandas as pd
```

```python
crime_data = pd.read_csv("../Resources/crime_data.csv")
crime_data
```

```python
#seperate columns 
years = crime_data.iloc[:,0]

total_violent_crime = crime_data.iloc[:,3]

murder_rate = crime_data.iloc[:,5]

aggravated_rape = crime_data.iloc[:,11]
```

```python
#linear regression 
lin1 = linregress(years, total_violent_crime)
 

lin2 = linregress(years, murder_rate)

lin3 = linregress(years, aggravated_rape)


#(slope, intercept, _, _, _) =  linregress(years, total_violent_crime)
#fit = slope * years + intercept
```

```python
(slope, intercept, _, _, _) = lin1
fit = slope + years + intercept
fig, ax = plt.subplots()

#(slope, intercept, _, _, _) = lin2
#fit = slope + years + intercept
#fig, ax = plt.subplots()

#(slope, intercept, _, _, _) = lin3
#fit = slope + years + intercept
#fig, ax = plt.subplots()

#Plot 

#fig, ax = plt.subplots()

fig.suptitle("Crime", fontsize=16, fontweight="bold")


ax.set_xlabel("years")
ax.set_ylabel("total_violent_crime")

ax.plot(years, total_violent_crime, linewidth=0, marker='o')
ax.plot(years, fit, 'b--')

plt.show()
```
